```python
# This code block performs data imputation, which means filling in missing values (NaN) in the dataset.
# It uses two common strategies:
# 1. For numerical columns, it replaces missing values with the mean (average) of that column.
#    This is a simple way to fill gaps while maintaining the overall average of the data.
# 2. For categorical columns, it replaces missing values with the mode (most frequent value) of that column.
#    This ensures that missing categories are filled with the most common category, which is a sensible default.

# Impute numerical columns with the mean
num_cols = ['Time_spent_Alone', 'Social_event_attendance', 'Going_outside',
            'Friends_circle_size', 'Post_frequency']
# For a column of numbers, if there are any empty spaces, it fills them with the average value of that column, and saves the change.
for col in num_cols:
    df[col].fillna(df[col].mean(), inplace=True)

# Impute categorical columns with the mode
cat_cols = ['Stage_fear', 'Drained_after_socializing']
# For a column of words or categories, if there are any empty spaces, it fills them with the most common word/category in that column, and saves the change.
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True) #[0] picks the first from the most repeated values if there is a tie.
```

<u>**üõ†Ô∏è Data Imputation Strategy**</u>

To ensure a complete dataset for analysis, we performed data imputation ‚Äî a process of filling in missing values using appropriate default values based on the column type.



<u>**üî¢ Numerical Columns**</u>

Missing values in numerical columns are replaced with the mean (average) of each column. This helps preserve the overall data distribution while ensuring no blanks disrupt calculations.



**<u>üî¢ Categorical Columns</u>**

For categorical data, missing values are filled with the mode (i.e., the most frequently occurring value). This ensures consistency by defaulting to the most typical category.



**<u>‚úÖ Why This Matters</u>**

- Prevents errors during model training and evaluation.

- Ensures statistical calculations aren't skewed by missing entries.

- Provides consistent, reproducible handling of null values.

<u>**üìä Exploratory Data Analysis (EDA)**</u>

The EDA process explores and visualizes patterns, distributions, and relationships within the dataset. Here's how the analysis is structured:

- Understand how individual features are spread across the dataset:

 - Box Plots: To detect outliers and observe the spread of data

<u>**üìà 2. Trend Analysis**</u>
- Identify patterns over time or index-based sequences:

- Line Charts: Useful if there‚Äôs a chronological or ordered aspect (e.g., how social activity changes across entries)

<u>**üîó 3. Correlation Analysis**</u>

- Explore relationships between variables:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the DataFrame to include only individuals categorized as 'Extrovert' (Personality_numeric == 1)
extrovert_df = df[df['Personality_numeric'] == 1]

#Checking the skewness of the data.
print("Skewness of Friends_circle_size (Extroverts):", extrovert_df['Friends_circle_size'].skew())

# Set the style for the plot
sns.set_style("white")

# This represents the width and height of the image.
plt.figure(figsize=(10, 7))

# Create a histogram of the 'Friends_circle_size' for Introverts.
# The bars are filled with a golden color and have brown borders, similar to your 'age' example.
# 'kde=False' means we are not adding a smooth line over the histogram.
ax = sns.histplot(data=extrovert_df, x='Friends_circle_size', kde=False,
                  color='gold', edgecolor='brown', linewidth=2.5)

# Add labels and title with custom fonts and size
plt.xlabel('Friends Circle Size', fontsize=14, fontweight='bold', labelpad=10)
plt.ylabel('Count', fontsize=14, fontweight='bold', labelpad=10)
plt.title('Distribution of Friends Circle Size for Extroverts',
          fontsize=16, fontweight='bold', pad=10)

# Add the number on top of each bar in the histogram
for p in ax.patches:
    height = p.get_height()
    if height > 0: # Only annotate bars that have a height greater than 0
        ax.annotate(f'{int(height)}',
                    (p.get_x() + p.get_width() / 2, height),
                    ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# Remove the right and top spines for a cleaner look.
sns.despine()

# Show the plot
plt.tight_layout()
plt.show()
```

**Skewness summary:**

Lastly, the skewness -0.25 means, the skew is mild (close to 0). The distribution is approximately symmetric, but with a slight tendency for lower values (on the left). This is typically not a strong skew and often may be considered close enough to normal for many practical purposes.


```python
# Creating a large figure (12 inches wide, 8 inches tall) to hold all the histograms
plt.figure(figsize=(12, 8))

# Create histograms for all numeric columns.
# - 30 bars per histogram (bins=30)
# - Arrange in 3x3 grid (layout=(3,3))
# - Big figure size for clarity (figsize=(12,8))
axes = df.hist(bins=30, figsize=(12, 8), layout=(3, 3))

# Flatten the 3x3 grid into a simple list
# Why? So we can loop through subplots one by one
axes = axes.flatten()

# Loop through each numeric column
# - 'i' = subplot position (0 to 8)
# - 'col' = column name (e.g., "Time spent Alone")
for i, col in enumerate(df.select_dtypes(include='number').columns):

    # Calculating skewness (how lopsided the data is)
    skewness = df[col].skew()

    # Adds title to each subplot:
    # - Column name + skewness value (2 decimal places)
    axes[i].set_title(f'{col}\nSkewness: {skewness:.2f}')

# Fixes spacing to avoid messy overlaps
plt.tight_layout()

# Shows the final result!
plt.show()
```

```python
# Calculate the first quartile (25th percentile) of the 'Friends_circle_size' column
Q1 = df['Friends_circle_size'].quantile(0.25)

# Calculate the third quartile (75th percentile) of the 'Friends_circle_size' column
Q3 = df['Friends_circle_size'].quantile(0.75)

# Compute the interquartile range (IQR)
IQR = Q3 - Q1

# Identify outliers using the IQR method
# An outlier is any value that lies below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
outliers = df[
    (df['Friends_circle_size'] < Q1 - 1.5 * IQR) |
    (df['Friends_circle_size'] > Q3 + 1.5 * IQR)
]

# Print the total number of outliers found in the 'Friends_circle_size' column
print(f"Outliers in Friends Circle Size: {len(outliers)}")
```

## Detecting Outliers in Friends Circle Size using the IQR Method

This code segment is used to detect outliers in the `Friends_circle_size` column of a dataset using the Interquartile Range (IQR) method.

### Steps Performed:
1. **Q1 and Q3 Calculation**:  
   The first (Q1) and third quartiles (Q3) of the `Friends_circle_size` column are computed to understand the distribution spread.

2. **IQR Calculation**:  
   The Interquartile Range (IQR) is found by subtracting Q1 from Q3.

3. **Outlier Detection**:  
   Any value below `Q1 - 1.5 * IQR` or above `Q3 + 1.5 * IQR` is considered an outlier. These values are filtered from the dataset.

4. **Result Output**:  
   The code prints the total count of such outliers detected.

### ‚úÖ Output:
**Outliers in Friends Circle Size: 0**

This means that all values in the `Friends_circle_size` column fall within the expected range, and there are no extreme or unusual data points based on the IQR method.


```python
from scipy.stats import ttest_ind  # Importing function for independent T-test

# Extracting 'Friends_circle_size' data for Extroverts
extroverts = df[df['Personality'] == 'Extrovert']['Friends_circle_size']

# Extracting 'Friends_circle_size' data for Introverts
introverts = df[df['Personality'] == 'Introvert']['Friends_circle_size']

# Performing an independent T-test to compare the means of two groups
t_stat, p_value = ttest_ind(extroverts, introverts)

# Displaying the T-test results
print(f"T-test result: t-stat = {t_stat}, p = {p_value}")
```

## üß™ Independent T-Test: Extroverts vs. Introverts

An independent **T-test** was conducted to determine whether there is a significant difference in the **Friends Circle Size** between Extroverts and Introverts.

### Steps Performed:
- The dataset was filtered to extract `Friends_circle_size` for:
  - Extroverts
  - Introverts
- The **`ttest_ind()`** function from `scipy.stats` was used to perform the test.
- The T-test compares the **means of two independent groups** to see if they are statistically different.

### üìä Result:
- **T-statistic** = `50.92`  
- **p-value** = `0.0`

### üìå Interpretation:
Since the **p-value is 0.0**, which is **less than 0.05**, we **reject the null hypothesis**.  
‚úÖ This indicates that there is a **statistically significant difference** in the average friends circle size between Extroverts and Introverts.

## üîç Chi-Square Test of Independence: Gender vs. Personality

A **Chi-square test of independence** was conducted to examine whether there is a significant association between **Gender** and **Personality Type** in the dataset.

### Steps Performed:
- A **contingency table** was created using `pd.crosstab()` to summarize the frequency distribution of:
  - Gender (e.g., Male, Female)
  - Personality Type (e.g., Introvert, Extrovert)
- The `chi2_contingency()` function from `scipy.stats` was used to perform the test.
- This test evaluates whether the observed distribution is significantly different from what we would expect if the variables were independent.

### üìä Result:
- **Chi-square value** = `1.35`

### üìå Interpretation:
